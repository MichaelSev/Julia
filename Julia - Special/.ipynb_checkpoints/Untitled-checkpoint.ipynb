#To calculate the y values, the loop is running from -5 to 5 



```julia
y = Vector{Int64,}()
for i in -5:5 
    push!(y,f(i))
    println("For x = $i is the value of y = $(y[i+6])")
end
```

#Exercise 6
#A plot to illustarte the values

```julia
using Plots 
gr()
plot(-5:5,y,
title = "Test of line",
xlabel = "X-value",
ylabel = "Y-value")
plot!(zero,-5,5,line=:scatter)
plot!(legend=false)
```

#The conclusion is that the function is zero at approximatly -2.2 and 2.2
