#Week 3, the SIR model

```julia
function updateSIR(popnvector)
    sus = popnvector[1];
    inf = popnvector[2];
    rem = popnvector[3];
    
    newS = sus - lambda * sus * inf * dt
    newI = inf + lambda * sus * inf * dt - gam * inf * dt
    newR = rem                           + gam * inf * dt
    
    return [newS newI newR]
    
end 


dt = 0.5
lambda = 1/200; gam = 1/10 

s, i, r = 1000., 10, 20 
vec = [s i r]

updateSIR(vec)
```
